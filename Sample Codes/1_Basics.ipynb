### Few-Shot

```python
examples = [
    {"input": "(2+2)*2", "output": "To solve this question \nStep 1:Higher priority will go to paranthesis so (2+2)=4 \nStep 2:Multiply by 2 so we will get answer 4*2 that is 8"},
    {"input": "(2*3)-2", "output": "To solve this question \nStep 1:Higher priority will go to paranthesis so (2*3)=6 \nStep 2:Subtract by 2 so we will get answer 6-2 that is 4"},
]
```

```python
# FewShots prompt
example_prompt = ChatPromptTemplate.from_messages(
    [
        ("human", "{input}"),
        ("ai", "{output}"),
    ]
)
few_shot_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
)

print(few_shot_prompt.format())
```

```python

final_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a wondrous wizard of math."),
        few_shot_prompt,
        ("human", user_input),
    ]
)
```

```python
print(final_prompt)
```

### Prompt Caching

Model prompts often contain repetitive content, like system prompts and common instructions. OpenAI routes API requests to servers that recently processed the same prompt, making it cheaper and faster than processing a prompt from scratch. Prompt Caching can reduce latency by up to 80% and input token costs by up to 90%. Prompt Caching works automatically on all your API requests (no code changes required) and has no additional fees associated with it. Prompt Caching is enabled for all recent models, gpt-4o and newer.

**Model on which it works**
1. gpt-5.2
2. gp5-5.1-codex-max
3. gpt-5.1
4. gpt-5.1-codex
5. gpt-5.1-codex-mini
6. gpt-5.1-chat-latest
7. gpt-5
8. gpt-5-codex
9. gpt-4.1

```python
client = OpenAI()
user_input = "How can i solve palindrome problem in python?"
completion = client.responses.create(
    model="gpt-5",
    prompt_cache_key="24h",
    input=[
        {"role": "system", "content": "You are an AI Assistant. Your task is to provide answer to each and every question of the user."},
        {
            "role": "user",
            "content": "I would like to visit the Eiffel Tower in Paris. Can you help me with some information about it?"
        }
    ],
)
print(completion.output_text)
```

### Custom Tool

```python
from langchain.chat_models import init_chat_model
from typing import Annotated
from typing_extensions import TypedDict
from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langchain_core.tools import tool
from langgraph.prebuilt import ToolNode, tools_condition
```

```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "path"
from dotenv import load_dotenv
load_dotenv()
```

```python

class State(TypedDict):
    # Messages have the type "list". The `add_messages` function
    # in the annotation defines how this state key should be updated
    # (in this case, it appends messages to the list, rather than overwriting them)
    messages: Annotated[list, add_messages]
```

```python
@tool
def get_stock_price(symbol: str) -> float:
    '''Return the current price of a stock given the stock symbol
    :param symbol: stock symbol
    :return: current price of the stock
    '''
    return{
        "MSFT":400,
        "APPL":300,
        "AMZN":500,
        "RIL": 250
    }.get(symbol,0.0)

tools = [get_stock_price]

llm = init_chat_model("google_genai:gemini-2.5-flash")
llm_with_tools = llm.bind_tools(tools)
```

```python
def LLM(state: State):
    return {"messages":[llm_with_tools.invoke(state['messages'])]}
```

```python
builder = StateGraph(State)

builder.add_node(LLM)
builder.add_node("tools",ToolNode(tools))

builder.add_edge(START,"LLM")
builder.add_conditional_edges("LLM",tools_condition)
builder.add_edge("LLM",END)

graph = builder.compile()
```

```python
graph.get_graph().print_ascii()
```

```python
state = graph.invoke({"messages":[{"role":"user","content":"What is the price of AMZN stock?"}]})
print(f"Answer: {state['messages'][-1].content}")
print(f"Working: {state}")   
```

```python
state = graph.invoke({"messages":[{"role":"user","content":"What is the sum of AMZN stock and MSFT stock?"}]})
print(f"Answer: {state['messages'][-1].content}")     # It will not give the sum just because it is calling the tool for getting the data and then again calling the tool for getting the data but doesnot call llm in between to understand the data and do the sum
print(f"Working: {state}")
```

```python
builder = StateGraph(State)

builder.add_node(LLM)
builder.add_node("tools",ToolNode(tools))

builder.add_edge(START, "LLM")

builder.add_conditional_edges(
    "LLM",
    tools_condition,   # if tool needed → tools
)

builder.add_edge("tools", "LLM")  # ← loop back

builder.add_edge("LLM", END)      # finish if no tool


graph = builder.compile()
```

```python
graph.get_graph().print_ascii()
```

```python
state = graph.invoke({"messages":[{"role":"user","content":"What is the sum of AMZN stock and MSFT stock?"}]})
print(f"Answer: {state['messages'][-1].content}")     # It will not give the sum just because it is calling the tool for getting the data and then again calling the tool for getting the data but doesnot call llm in between to understand the data and do the sum
print(f"Working: {state}")
```
